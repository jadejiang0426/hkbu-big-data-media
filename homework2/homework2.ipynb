```python
import requests
import bs4
import csv
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://theinitium.com/channel/news-brief/')
```

```python
r.text
```

```python
document = BeautifulSoup(r.text)
```

```python
document
```

```python
titles = document.find_all ('h3',attrs={'class':'c-digest-headline c-digest-headline--sm'})
```

```python
titles
```

```python
list_title = []
for title in titles:
    title.text
    list_title.append(title.text)
```

```python
list_title
```

```python
datetimes = document.find_all ('time',attrs={'class':'c-digest-date u-font-sans'})
```

```python
datetimes
```

```python
list_datetime = []
for datetime in datetimes:
    datetime.text
    list_datetime.append(datetime.text)
```

```python
list_datetime
```

```python
with open ('homework2.csv','w') as f:
    writer = csv.writer(f)
    writer.writerows(zip(list_title,list_datetime))
```
